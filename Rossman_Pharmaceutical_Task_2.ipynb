```python
def handle_outliers_median(df, columns_to_check):
    for column in columns_to_check:
        Q1 = np.percentile(df[column], 25)
        Q3 = np.percentile(df[column], 75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers_mask = (df[column] < lower_bound) | (df[column] > upper_bound)
        df.loc[outliers_mask, column] = np.median(df[column])  # Replace outliers with median
    return df

# Numeric columns
numeric_cols = train_store_combine_data.select_dtypes(include=np.number).columns.tolist()

# Handle outliers and replace with median
train_store_combine_data_handled = handle_outliers_median(train_store_combine_data.copy(), numeric_cols)

# Loop through columns and create appropriate plots
for col in train_store_combine_data_handled.columns:
    if train_store_combine_data_handled[col].dtype != 'object' and train_store_combine_data_handled[col].dtype != 'datetime64[ns]':
        plt.figure(figsize=(6, 4))
        sns.boxplot(data=train_store_combine_data_handled[col])
        plt.title(f'Box Plot - {col}')
        plt.show()
    elif train_store_combine_data_handled[col].dtype == 'object':
        plt.figure(figsize=(6, 4))
        sns.countplot(data=train_store_combine_data_handled, x=col)
        plt.title(f'Count Plot - {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.savefig(f'count_{col}.png')
        plt.show()
```

```python
from pmdarima import auto_arima
from statsmodels.tsa.stattools import adfuller


print(train_store_combine_data.dtypes)


# Downsample the 'Sales' column to monthly frequency
monthly_sales = train_store_combine_data.resample('M')['SalesPerCustomer'].sum()

# Define a function to perform ADF test for stationarity
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    p_value = result[1]
    return p_value <= 0.05

# Check if differencing is needed
if adf_test(monthly_sales):
    d = 0
else:
    d = 1

# Use auto_arima for automatic ARIMA order selection
model = auto_arima(monthly_sales, start_p=1, start_q=1,
                   max_p=3, max_q=3, m=12,
                   start_P=0, seasonal=True,
                   d=d, D=1, trace=True,
                   error_action='ignore',
                   suppress_warnings=True,
                   stepwise=True)

# Fit the ARIMA model
model_fit = model.fit(monthly_sales)

# Print the model summary
print(model_fit.summary())
```
